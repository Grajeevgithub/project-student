```python

# Load the data
data = pd.read_csv('Orange Quality Data.csv')

# Clean column names
data.columns = data.columns.str.strip()

# Sort by Harvest Time for proper sequencing
data = data.sort_values(by='HarvestTime (days)')

# Plot
plt.figure(figsize=(10, 6))
plt.plot(data['HarvestTime (days)'], data['Brix (Sweetness)'],
         marker='o', linestyle='-', color='darkorange')
plt.title('Sweetness (Brix) vs. Harvest Time')
plt.xlabel('Harvest Time (days)')
plt.ylabel('Brix (Sweetness)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# barplot

```python

# Load the dataset
data = pd.read_csv('Orange Quality Data.csv')

# Clean and strip column names
data.columns = data.columns.str.strip()

# Group by variety and calculate average quality
avg_quality = data.groupby('Variety')['Quality (1-5)'].mean().sort_values()

# Plot
plt.figure(figsize=(12, 6))
avg_quality.plot(kind='barh', color='darkorange', edgecolor='black')
plt.title('Average Quality by Orange Variety')
plt.xlabel('Average Quality (1–5)')
plt.ylabel('Variety')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit

```python
import pandas as pd

# Load and clean
data = pd.read_csv('Orange Quality Data.csv')
data.columns = data.columns.str.strip()

# Calculate correlation
correlation = data['Size (cm)'].corr(data['Weight (g)'])
print(f"Correlation between Size and Weight: {correlation:.2f}")
```

```python

# Scatter plot with regression line
sns.lmplot(x='Size (cm)', y='Weight (g)', data=data, height=6, aspect=1.2,
           scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
```

# conclusion
* The visualization confirmed a strong positive correlation between size and weight of the oranges — meaning larger oranges tend to weigh more, as expected. The scatter plot with a regression line visually reinforced this trend, showing a clear upward pattern. This insight supports intuitive expectations and validates the consistency of measurement in your dataset.

# How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity

```python


# Calculate correlation
correlation = data['Brix (Sweetness)'].corr(data['pH (Acidity)'])
print(f"Correlation between Brix and pH: {correlation:.2f}")
```

```python

# Scatter plot with regression line
sns.lmplot(x='Brix (Sweetness)', y='pH (Acidity)', data=data, height=6, aspect=1.2,
           scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
```

# conclusion 
* The visual and statistical analysis shows a moderate negative correlation between sweetness (Brix) and acidity (pH) — meaning that, overall, sweeter oranges tend to be slightly less acidic. This aligns with expectations from citrus ripening patterns: as sugars increase, acidity typically mellows out.
* The regression line in the scatter plot sloped downward, reinforcing this inverse relationship. However, since it's not a perfectly strong correlation, other factors like variety, ripeness stage, or environmental conditions may also influence these properties.

* Depending on the strength of the correlation and the spread in your plots (box or violin), you might also spot subtle variations influenced by factors like variety or harvest time. Overall, softness appears to be a reliable indicator of ripeness in your dataset — a juicy insight for quality assessment.

# How do different fruit varieties (Variety) differ in terms of size, weight, sweetness, acidity, and ripeness

```python
import pandas as pd

# Load and clean
data = pd.read_csv('Orange Quality Data.csv')
data.columns = data.columns.str.strip()

# Group by variety and calculate mean values
summary = data.groupby('Variety')[['Size (cm)', 'Weight (g)', 'Brix (Sweetness)', 'pH (Acidity)', 'Ripeness (1-5)']].mean().round(2)

# Sort by sweetness for example
summary = summary.sort_values(by='Brix (Sweetness)', ascending=False)

print(summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Sweetness by Variety
sns.boxplot(x='Variety', y='Brix (Sweetness)', data=data)
plt.xticks(rotation=90)
```

```python
sns.violinplot(x='Variety', y='pH (Acidity)', data=data)
plt.xticks(rotation=90)
```

```python
summary = data.groupby('Variety')[['Size (cm)', 'Weight (g)', 'Brix (Sweetness)', 'pH (Acidity)', 'Ripeness (1-5)']].mean()
sns.heatmap(summary, annot=True, cmap='YlOrRd')
```

# conclusion
* Different orange varieties exhibit distinct profiles across size, weight, sweetness, acidity, and ripeness. Some varieties (like Cara Cara and Satsuma Mandarin) may consistently score higher on sweetness (Brix), while others (like Moro or Valencia) may lean toward more acidity or larger size.
* If your visualizations — such as box plots or heatmaps — showed clear separation between groups, it suggests that variety is a meaningful predictor of key fruit attributes. This insight is valuable for both growers (targeting flavor or size traits) and marketers (emphasizing consumer preferences).
