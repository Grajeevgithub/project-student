# lineplot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('sentimentdataset.csv')

# Convert Timestamp to datetime object
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Group by date and sentiment
daily_counts = data.groupby([data['Timestamp'].dt.date, 'Sentiment']).size().unstack(fill_value=0)

# Plot the lineplot
plt.figure(figsize=(14, 6))
daily_counts.plot(kind='line', marker='o', linewidth=2)
plt.title('Sentiment Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Posts')
plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# barplot

```python
sentiment_counts = data['Sentiment'].value_counts()

# Plot as bar chart
plt.figure(figsize=(10, 6))
sentiment_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Post Count by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Number of Posts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

# scatterplot

# What is the overall distribution of sentiments (positive, negative, neutral) across the posts

```python
# Define mapping to Positive, Negative, and Neutral
sentiment_map = {
    'Positive': 'Positive', 'Negative': 'Negative', 'Neutral': 'Neutral',
    'Joy': 'Positive', 'Love': 'Positive', 'Gratitude': 'Positive', 'Happiness': 'Positive',
    'Excitement': 'Positive', 'Admiration': 'Positive', 'Pride': 'Positive',
    'Amusement': 'Positive', 'Enjoyment': 'Positive', 'Contentment': 'Positive',
    'Affection': 'Positive', 'Awe': 'Positive', 'Surprise': 'Positive', 'Hope': 'Positive',
    'Enthusiasm': 'Positive', 'Inspiration': 'Positive', 'Confidence': 'Positive',
    'Curiosity': 'Positive', 'Kind': 'Positive', 'Calmness': 'Positive',

    'Sadness': 'Negative', 'Anger': 'Negative', 'Fear': 'Negative', 'Disgust': 'Negative',
    'Shame': 'Negative', 'Bitterness': 'Negative', 'Frustration': 'Negative',
    'Jealousy': 'Negative', 'Resentment': 'Negative', 'Loneliness': 'Negative',
    'Regret': 'Negative', 'Despair': 'Negative', 'Hate': 'Negative', 'Grief': 'Negative',
    'Anxiety': 'Negative', 'Apprehensive': 'Negative', 'Disappointed': 'Negative',
    'Helplessness': 'Negative', 'Envy': 'Negative', 'Devastated': 'Negative',
    'Bad': 'Negative', 'Heartbreak': 'Negative', 'Loss': 'Negative', 'Desolation': 'Negative',

    # Catch any unmapped entries
    'Happy': 'Positive', 'Sad': 'Negative'
}

# Apply the mapping
data['BroadSentiment'] = data['Sentiment'].map(sentiment_map).fillna('Other')
```

```python
import matplotlib.pyplot as plt

# Count sentiments
sentiment_counts = data['BroadSentiment'].value_counts()

# Plot using one color or colormap to avoid color mismatch
plt.figure(figsize=(10, 6))
sentiment_counts.plot(kind='bar', color='cornflowerblue', edgecolor='black')
plt.title('Overall Distribution of Sentiments')
plt.xlabel('Sentiment Category')
plt.ylabel('Number of Posts')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# conclusion
* Positive sentiments dominate the dataset â€” the majority of posts express emotions like joy, gratitude, pride, excitement, and love. This suggests that users are largely using these platforms to share uplifting and emotionally expressive moments.
* Negative sentiments are present in meaningful volume, encompassing feelings like sadness, anger, fear, frustration, and regret. These reflect emotional vulnerability or reactions to challenging events but are notably less common than positive ones.
* Neutral sentiments make up a smaller slice of the total, mostly tied to informational, observational, or less emotionally charged posts.

#  Which platform generates the most engagement in terms of likes and retweets

```python


# Load dataset if not already loaded
# data = pd.read_csv('sentimentdataset.csv')

# Group by platform and calculate average likes and retweets
engagement_stats = data.groupby('Platform')[['Likes', 'Retweets']].mean().sort_values(by='Likes', ascending=False)

# View results
print(engagement_stats)
```

```python

# Plot average engagement by platform
engagement_stats.plot(kind='bar', figsize=(10, 6), color=['mediumseagreen', 'cornflowerblue'], edgecolor='black')
plt.title('Average Likes and Retweets by Platform')
plt.ylabel('Average Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# conclusion
* Based on the engagement analysis, Twitter generates the highest average likes and retweets, making it the most interactive platform in this dataset. Its design emphasizes rapid sharing and public reactions, which naturally boosts engagement compared to platforms like Facebook and Instagram. This highlights Twitter's strength as a hub for emotional expression and viral content diffusion.

#  Which users have the highest average engagement (likes and retweets per post)

```python
# Group by user and calculate average likes and retweets
user_engagement = data.groupby('User')[['Likes', 'Retweets']].mean()

# Add a new column for total average engagement
user_engagement['AvgEngagement'] = user_engagement['Likes'] + user_engagement['Retweets']

# Sort by highest average engagement
top_users = user_engagement.sort_values(by='AvgEngagement', ascending=False)

# Display top 10 users
print(top_users.head(10))
```

```python


# Step 1: Group and calculate average engagement
user_stats = data.groupby('User')[['Likes', 'Retweets']].mean()
user_stats['AvgEngagement'] = user_stats['Likes'] + user_stats['Retweets']

# Step 2: Sort and pick top users
top_users = user_stats.sort_values(by='AvgEngagement', ascending=False).head(10)

# Step 3: Plot
plt.figure(figsize=(12, 6))
top_users['AvgEngagement'].plot(kind='barh', color='teal', edgecolor='black')
plt.title('Top 10 Users by Average Engagement (Likes + Retweets)')
plt.xlabel('Average Engagement per Post')
plt.ylabel('User')
plt.gca().invert_yaxis()  # Highest at top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# conclusion
* The analysis revealed that the users with the **highest average engagement** (likes + retweets per post) are standout voices in the dataset, consistently capturing audience attention. These users likely share compelling content, post at optimal times, or maintain a strong follower baseâ€”making them key influencers in their respective platforms and valuable for trendsetting or campaign amplification.
