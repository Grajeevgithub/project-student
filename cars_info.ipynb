# lineplot


```python
avg_price_by_year = data.groupby('Year')['Price'].mean().reset_index()

# Plot
sns.lineplot(data=avg_price_by_year, x='Year', y='Price', marker='o')
plt.title('Average Car Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# scatterplot

```python
# Basic scatterplot of Price vs Kilometers
sns.scatterplot(data=data, x='Kilometers', y='Price')
plt.title('Price vs Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

# boxplot

```python
sns.boxplot(data=data, x='Fuel', y='Price')
plt.title('Car Price Distribution by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# histoplot

# barplot

```python
avg_price_by_fuel = data.groupby('Fuel')['Price'].mean().reset_index()

# Plot
sns.barplot(data=avg_price_by_fuel, x='Fuel', y='Price')
plt.title('Average Car Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Is there a correlation between mileage (Kilometers) and the price of a car

### It is related lets check with the scatter plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=data, x='Kilometers', y='Price', alpha=0.6)
plt.title('Car Price vs. Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
sns.regplot(data=data, x='Kilometers', y='Price', scatter_kws={'alpha':0.4})
plt.title('Regression: Price vs. Kilometers')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

# Conclusion: Mileage vs. Car Price
* There is a negative correlation between mileage (Kilometers driven) and car price. As the number of kilometers increases, the price of the car tends to decrease. This trend suggests that buyers perceive higher mileage as an indicator of greater wear and tear or reduced remaining lifespan, thereby lowering the car's market value.
* While exceptions may exist due to factors like brand, condition, or maintenance history, the overall pattern confirms that mileage is a key determinant in car pricingespecially in the used car market.
