# lineplot


```python

avg_purchase = data.groupby('Product Category')['Total Purchase Amount'].mean().reset_index()

sns.lineplot(data=avg_purchase, x='Product Category', y='Total Purchase Amount', marker='o')
plt.title('Average Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Avg Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# scatter plot

```python
sns.scatterplot(data=data, x='Customer Age', y='Total Purchase Amount', alpha=0.6)
plt.title('Total Purchase Amount vs Customer Age')
plt.xlabel('Customer Age')
plt.ylabel('Total Purchase Amount')
plt.tight_layout()
plt.show()
```

 # What are the trends in total purchase amount over time (based on Purchase Date)? Are there any seasonal patterns

```python
data['Purchase Date'] = pd.to_datetime(data['Purchase Date'])
```

```python
monthly_sales = data.groupby(data['Purchase Date'].dt.to_period('M'))['Total Purchase Amount'].sum().reset_index()
monthly_sales['Purchase Date'] = monthly_sales['Purchase Date'].dt.to_timestamp()
```

```python
sns.lineplot(data=monthly_sales, x='Purchase Date', y='Total Purchase Amount', marker='o')
plt.title('Monthly Total Purchase Amount Over Time')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion: Temporal Trends and Seasonality in Purchase Behavior
* The time series analysis of total purchase amount reveals clear temporal trends and potential seasonal patterns in customer behavior. Periodic peaks—such as year-end surges or mid-year slumps—may align with promotional campaigns, holidays, or fiscal cycles. A generally rising or declining trend could reflect broader shifts in customer engagement or market dynamics.
* These patterns suggest that purchase behavior is influenced by both time and context, offering valuable insights for inventory planning, marketing timing, and revenue forecasting. Identifying and leveraging seasonal upticks can help businesses optimize conversion and retention strategies.

# Who are the top customers by total purchase amount or frequency, based on Customer ID and Customer Name

```python
top_spenders = data.groupby(['Customer ID', 'Customer Name']).apply(
    lambda x: (x['Product Price'] * x['Quantity']).sum()
).reset_index(name='Total Purchase Amount')
```
