# 1) How does the number of hours studied per week correlate with the final exam score?

```python

plt.figure(figsize=(10,8))
plt.scatter(data['Hours_Studied'], data['Exam_Score'], color='blue',)
plt.title('Hours Studied vs. Final Exam Score')
plt.xlabel('Hours Studied per Week')
plt.ylabel('Final Exam Score')
plt.show()
```

# 2) Is there a significant relationship between class attendance percentage and exam scores?

```python

data['Attendance'] = pd.to_numeric(data['Attendance'])
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data[['Attendance','Exam_Score']].corr(), annot=True, cmap='coolwarm')
plt.title("Relation btw attendance percentage and exam scores")
plt.show()
```

# 3) Does the average number of sleep hours per night influence exam performance?

```python
data['Sleep_Hours'].mean()
```

# 4) How does family income and parental education level impact the student's exam scores?

```python
data = data.groupby(['Family_Income','Parental_Education_Level'])['Exam_Score'].mean()

     
```

```python
data
```

# 5.Do private school students perform better in exams than public school students?

```python
sns.boxplot(x='School_Type', y='Exam_Score', data=data)
plt.title("Exam Score Distribution by School Type")
plt.show()
```

```python
sns.histplot(data=data, x='Exam_Score', hue='School_Type', kde=True, element='step')
plt.title("Distribution of Exam Scores by School Type")
plt.show()
```

```python
avg_scores = data.groupby('School_Type')['Exam_Score'].mean().reset_index()
sns.barplot(x='School_Type', y='Exam_Score', data=avg_scores)
plt.title("Average Exam Scores by School Type")
plt.show()
```

```python
sns.countplot(x='School_Type', data=data)
plt.title("Number of Students by School Type")
plt.show()
```

```python
sns.scatterplot(x='Hours_Studied', y='Exam_Score', hue='School_Type', data=data)
plt.title("Hours Studied vs Exam Score by School Type")
plt.show()
```

```python
numeric_data = data.select_dtypes(include='number')
corr = numeric_data.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
